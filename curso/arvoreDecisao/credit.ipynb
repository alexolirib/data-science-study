```python
# importação de arquivo
import pandas as pd
import numpy as np
# visualização de gráficos
import seaborn as sns
# visualização de gráficos
import matplotlib.pyplot as plt
# visualizar grafico dinâmico 
import plotly.express as px
# GaussianNB - Nome de uma distirbuição estatistica (usado para problemas genericos
from sklearn.tree import DecisionTreeClassifier
import pickle
```

```python
with open('./../base/credit.pkl', 'rb') as f:
    X_credit_treinamento, y_credit_treinamento, X_credit_test, y_credit_test = pickle.load(f)
```

```python
X_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
X_credit_test.shape, y_credit_test.shape
```

```python
tree_credit = DecisionTreeClassifier(criterion='entropy', random_state=0)
tree_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = tree_credit.predict(X_credit_test)
previsoes
```

```python
y_credit_test
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_credit_test, previsoes)
```

```python
# matriz de confusão
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(tree_credit)
cm.fit(X_credit_treinamento, y_credit_treinamento)
cm.score(X_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, previsoes))
```

```python
tree_credit.classes_
```

```python
# visualizar a arvore de decisao
from sklearn import tree
previsores = ['income', 'age', 'loan']
figura, eixos = plt.subplots(nrows=1, ncols=1, figsize=(20,20))
# 0 -> paga
# 1 ->  nao paga
tree.plot_tree(tree_credit, feature_names=previsores, class_names=['0', '1'], filled=True);
figura.savefig('tree_credit.png')
```
